```python
# Import tokenizers and transformers library
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
# Call your tokenizer directly on the sentence, you get back inputs that are ready to pass through your model
checkpoint = 'gpt2'
tokenizer = AutoTokenizer.from_pretrained('gpt2')

sequence = "I've been waiting for a Hugging Face course my whole life."

model_inputs = tokenizer(sequence, return_tensors='pt') # 'pt' means returns PyTorch tensors, alternatively, 'tf' returns TensorFlow tensor, and 'np' returns Numpy arrays
model_inputs
```

```python
# Here the 'model_inputs' parameter is the only necessary for a model, and the 'attention_mask' is optional for multiple sequences handling.
model = AutoModelForCausalLM.from_pretrained('gpt2')
model(**model_inputs)
```

Bam! That's all your need for putting tokenizer and transformer model together and make prediction.
